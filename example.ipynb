```python
from feature import mfcc
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
import librosa
```

```python
PATH = 'AudioData/'
sr, audio = wavfile.read(PATH + "Output_0.wav")
print(type(audio))
print("Sample rate: {0}Hz".format(sr))
print("Audio duration: {0}s".format(len(audio) / sr))
print(type(audio))
```

```python
plt.figure(figsize=(26,4))
for i in range(5):
    sr, data = wavfile.read(PATH + "Output_"+str(i)+".wav")
    coefs, power = mfcc(data, sr)

    plt.subplot(1, 5, (i+1))
    plt.plot(np.linspace(0, len(audio) / sr, num=len(audio)), audio)
    plt.imshow(power, aspect='auto', origin='lower')
```

```python
coefs, power = mfcc(audio, sr, n_fft=1024, hop_length=200, n_mfcc=20, n_filters=60)

data = librosa.feature.mfcc(y=audio, sr=sr, n_fft=1024, hop_length=200, n_mfcc=20)


plt.figure(figsize=(10,10))
plt.subplot(2, 1, 1)
plt.title("My lib")
plt.plot(np.linspace(0, len(audio) / sr, num=len(audio)), audio)
plt.imshow(coefs, aspect='auto', origin='lower')
plt.subplot(2, 1, 2)
plt.title("Liborsa")
plt.plot(np.linspace(0, len(audio) / sr, num=len(audio)), audio)
plt.imshow(data, aspect='auto', origin='lower')
```
